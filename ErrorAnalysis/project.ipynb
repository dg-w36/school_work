```python
from scipy.optimize import leastsq  # 引入最小二乘函数  

# 多项式函数  
def fit_func(p, x):  
    f = np.poly1d(p)  
    return f(x)  

# 残差函数  
def error_func(p, y, x):  
    ret = fit_func(p, x) - y  
    return ret  

p_init = np.random.randn(2)
plsq = leastsq(error_func, p_init, args=(Y, X))

print plsq[0]
print "function is: y = %lfx + %lf"%(plsq[0][0], plsq[0][1])
```
