# 误差的基本概念和误差分析
## 误差的定义
测量结果－测量真值：
$$
\varepsilon = x - x_0
$$
值得注意的是，误差是存在正负的，并不是恒正。
### 真值的获取
对于测量来说，未知情况下，很难得到理论的测量真值，通常用统计方法计算得到真值的估计。其它情况下真值可以由理论真值取得，或者取计量学约定真值。
### 几种实际使用时的误差
按照误差的表示形式分：
绝对误差：与上述误差定义一致，测量值－真值  
相对误差：绝对误差与真值的比值（$\varepsilon_r = \frac{\varepsilon}{x_0}$）可以用来反映测量精度  
引用误差：使用测量仪器测量时，测量数值所带有的误差  
按照误差的性质分：
系统误差：系统误差可以看做由测量系统引入，测量量整体性偏离，有一定趋势  
随机误差：误差存在随机的波动，展现出单峰性，抵偿性的特点  
粗大误差：某一次测试数据误差显著大于其它误差，显示为实验的操作失误造成  

### 绝对误差
$\delta_x = X - X_0$
测量值－真值  
绝对误差有具体确定大小，符号，单位
**修正值**：为了消除固定的系统误差用代数法加到数据上

### 相对误差
$\displaystyle r = \frac{\delta_x}{X_0}$
有符号和大小 无量纲 用百分比表示  
相对误差用于比较测量的精度

### 引用误差
量程内，最大绝对误差与测量范围上线或满量程之比，一般是测量仪器带来的误差，引起误差由引用误差与量程乘积计算。

## 数据有效数字与数值计算
根据数据的精度，决定数据的位数，数据中间处理量，要比有效数字多一位可疑位。
### 数据修约原则
四舍五入，要求后面大于“五”，若恰好为五，有留双的原则。
0.3351 -> 0.34 (大于“5”)
0.3350 -> 0.34（等于5，留双）
0.3450 -> 0.34（等于5，留双）
**不能连续修约**

### 数据运算的有效数字要求
总体规则：先将数据修约到有效数字多一位，再进行计算，修约到有效数字位数。
#### 加减运算
**小数位数**最少的数据位，为有效数字。
13.65 ＋ 0.0082 ＋ 1.637 (13.65位数最少，则有效数字2位)
13.65 ＋ 0.008 ＋ 1.637 （修约到多一位可疑位）
15.295 -> 15.30
#### 乘除运算
**有效数字最少**的数据的有效数字位为标准。
1.3048 * 236 （选236，有效为3位）
1.305 * 236 （修约）
307.98 -> 308
乘方运算和开方运算是相同的。
