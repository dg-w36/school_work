```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%pylab inline
```

```python
img = cv2.imread('./data/1.jpg',1)
gray = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY)
imshow(gray, cmap='gray')
# plt.subplot(222); plt.imshow(img, cmap ='gray')
```

```python
blur = cv2.blur(gray,(5,5))
imshow(blur, cmap='gray')
```

```python
rst, thresh = cv2.threshold(blur, 127, 255, cv2.THRESH_TOZERO)
imshow(thresh, cmap='gray')
```

```python
img2 = img.copy()
image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
img3 = cv2.drawContours(img2, contours, 1, (0,255,0), 3)
imshow(img3)
```

```python
x,dim,y = contours[1].shape
contours.sort(key=lambda tmp:[tmp[0:x,0,0].mean(), tmp[0:x,0,1].mean()])
```

```python
img2 = img.copy()
img3 = cv2.drawContours(img2, contours, -1, (0,255,0), 3)
imshow(img3)
```

```python
img2 = img.copy()
for cnt in contours :
    x,y,w,h = cv2.boundingRect(cnt)
    img3 = cv2.rectangle(img2,(x,y),(x+w,y+h),(255,0,0),2)
imshow(img3)
```
